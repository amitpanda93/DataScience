```python
#Import the packages

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, recall_score, precision_score

# Divide data into dependent and independent variables

# Over sampled dataset
# x = os_train.drop("Target", axis = 1)
# y = os_train.Target

#Under sampled datset
# x = us_train.drop("Target", axis = 1)
# y = us_train.Target



x = train_data_new.drop("Target", axis = 1)
y = train_data_new.Target

#Scale the data
x_sc = pd.DataFrame(MinMaxScaler().fit_transform(x), columns=x.columns)

#Split the data into train and test data

x_train, x_test, y_train, y_test = train_test_split(x_sc,y,test_size = 0.33,random_state = 75)


# Fit the Logistic Regression model

lr = LogisticRegression().fit(x_train,y_train)

#Predicting on train values
train_pred = lr.predict(x_train)

#Predicting on test values
test_pred = lr.predict(x_test)

#Print the ROC score
print("ROC score: ", roc_auc_score(y_test, test_pred))


#print the confusion matrix

print("-------------")
print("Confusion matrix for train: \n", confusion_matrix(y_train, train_pred, labels = [0,1]))
print("-------------")
print("Confusion matrix for test: \n", confusion_matrix(y_test, test_pred, labels = [0,1]))

#print count of true positive, true negative, false positive, false negative

tn, fp, fn, tp = confusion_matrix(y_train, train_pred).ravel()
print("-------------")
print("True Negative: ", tn , "\nFalse Positive: ", fp ,"\nFalse Negative: ", fn, "\nTrue Positive: ",tp)

#Print accuracy of the model
print("-------------")

print("Accuracy on train: ", accuracy_score(y_train, train_pred))
print("Accuracy on test: ", accuracy_score(y_test, test_pred))

#Print recall of the model
print("-------------")

print("Recall on train: ", recall_score(y_train, train_pred))
print("Recall on test: ", recall_score(y_test, test_pred))

#Print precision of the model
print("-------------")

print("Precision on train: ", precision_score(y_train, train_pred))
print("Precision on test: ", precision_score(y_test, test_pred))

print("-------------")
```
